```python
import pandas as pd
from datetime import datetime
```

```python
start_date="2015-01-01"
end_date="2025-01-01"
```

```python
df1=pd.DataFrame()
```

```python
def function(df,start_date,end_date,format):
    df=df.copy()
    df["Tarih"]=pd.to_datetime(df["Tarih"],format=format)
    df.set_index("Tarih",inplace=True)
    df.sort_index(inplace=True)
    df=df.loc[start_date:end_date].resample('D').interpolate(method='linear')
    df=df[start_date:end_date]
    return df
```

```python
dfTHYAO=pd.read_csv('THYAO.csv',sep=',',decimal=',')
dfTHYAO=dfTHYAO.iloc[::-1].reset_index(drop=True)
formatTHYAO='%d.%m.%Y'
dfTHYAO=function(dfTHYAO,start_date,end_date,formatTHYAO)
dfTHYAO.tail()
```

```python
dfM1=pd.read_csv('M1.csv',sep=',',decimal='.')
formatM1='%Y-%d-%m'
dfM1=function(dfM1,start_date,end_date,formatM1)
dfM1.tail()
```

```python
dfInflation=pd.read_csv('Inflation.csv',sep=',',decimal='.')
formatInflation='%Y-%d-%m'
df1Inflation=function(dfInflation,start_date,end_date,formatInflation)
df1Inflation.tail()
```

```python
dfGDP=pd.read_csv('GDP.csv',sep=',',decimal='.')
formatGDP='%Y-%d-%m'
dfGDP=function(dfGDP,start_date,end_date,formatGDP)
dfGDP.tail()
```

```python
dfBIST100=pd.read_csv('BIST100.csv',sep=',',decimal=',',thousands='.')
formatBIST100='%Y-%m-%d'
dfBIST100=function(dfBIST100,start_date,end_date,formatBIST100)
dfBIST100.tail()
```

```python
dfUSDTRY=pd.read_csv('USDTRY.csv',sep=',',decimal=',')
formatUSDTRY='%Y-%m-%d'
dfUSDTRY=function(dfUSDTRY,start_date,end_date,formatUSDTRY)
dfUSDTRY.tail()
```

```python
dfInterestRate=pd.read_csv('InterestRate.csv',sep=',',decimal='.')
formatInterestRate='%Y-%m-%d'
dfInterestRate=function(dfInterestRate,'2014-08-28',end_date,formatInterestRate)
dfInterestRate.tail()
```

```python
merged_df=dfTHYAO.join([dfM1,df1Inflation,dfGDP,dfInterestRate,dfUSDTRY,dfBIST100],how="inner")
#merged_df.loc[start_date:end_date]
merged_df=merged_df.round(2)
merged_df.tail()
```

```python
merged_df.to_csv("Merged_Data.csv")
```
